```python
import sounddevice as sd
import numpy as np
import tkinter as tk
from tkinter import messagebox
from tkinter import simpledialog
from IPython.display import Audio, display
from tensorflow.keras.models import load_model
import pyttsx3
import librosa
import pygame

# Load the model
model = load_model('model3.h5')

# Constants for audio recording
duration = 3  # Duration of recording in seconds
sample_rate = 22050  # Sample rate of audio

# Constants for character animation
CHARACTER_WIDTH = 300
CHARACTER_HEIGHT = 300

# Initialize Pygame
pygame.init()

# Load character sprite image
character_image = pygame.image.load(r"C:\Users\Acer pc\Downloads\download.jpg")
character_image = pygame.transform.scale(character_image, (CHARACTER_WIDTH, CHARACTER_HEIGHT))

# Set up Pygame window
window_width = 500
window_height = 500
window = pygame.display.set_mode((window_width, window_height))

# Function to record audio
def record_audio():
    messagebox.showinfo("Recording", "Recording started. Please speak into the microphone for {} seconds.".format(duration))
    audio_data = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1, blocking=True)
    messagebox.showinfo("Recording", "Recording finished.")
    return audio_data.flatten()

# Function to extract audio features
def extract_feature(audio_data, mfcc, chroma, mel):
    if chroma:
        stft = np.abs(librosa.stft(audio_data))
    result = np.array([])
    if mfcc:
        mfccs = librosa.feature.mfcc(y=audio_data, sr=sample_rate, n_mfcc=40)
        mfccs_mean = np.mean(mfccs, axis=1)
        result = np.hstack((result, mfccs_mean))
    if chroma:
        chroma = librosa.feature.chroma_stft(S=stft, sr=sample_rate)
        chroma_mean = np.mean(chroma, axis=1)
        result = np.hstack((result, chroma_mean))
    if mel:
        mel = librosa.feature.melspectrogram(y=audio_data, sr=sample_rate)
        mel_mean = np.mean(mel, axis=1)
        result = np.hstack((result, mel_mean))
    return result

# Function to predict emotion from audio
def predict_emotion():
    audio_data = record_audio()
    features = extract_feature(audio_data, mfcc=True, chroma=True, mel=True)
    prediction = model.predict(np.expand_dims(features, axis=0))
    predicted_class = np.argmax(prediction)
    emotion_mapping = {0: 'Angry', 1: 'Happy', 2: 'Sad', 3: 'Neutral'}
    predicted_emotion = emotion_mapping[predicted_class]
    messagebox.showinfo("Prediction", "Predicted Emotion: {}".format(predicted_emotion))
    display(Audio(audio_data, rate=sample_rate))

    # Generate personalized audio greeting
    engine = pyttsx3.init()
    engine.say(f"Hello {name}. I detected {predicted_emotion} in your voice. Here is the audio you recorded.")
    engine.runAndWait()

    # Play the recorded audio
    display(Audio(audio_data, rate=sample_rate))

    # Animate character speaking
    character_speak()

# Function to animate character speaking
def character_speak():
    # Create a surface for the character image
    character_surface = pygame.Surface((CHARACTER_WIDTH, CHARACTER_HEIGHT), pygame.SRCALPHA)

    # Blit the character image onto the surface
    character_surface.blit(character_image, (0, 0))

    # Draw a speech bubble
    speech_bubble_rect = pygame.Rect(CHARACTER_WIDTH + 10, CHARACTER_HEIGHT - 100, 200, 80)
    pygame.draw.rect(character_surface, (255, 255, 255), speech_bubble_rect)
    pygame.draw.rect(character_surface, (0, 0, 0), speech_bubble_rect, 2)

    # Create a font object
    font = pygame.font.SysFont(None, 24)

    # Render the text onto the surface
    text = font.render("I'm speaking!", True, (0, 0, 0))
    text_rect = text.get_rect(center=speech_bubble_rect.center)
    character_surface.blit(text, text_rect)

    # Main game loop
    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Clear the window
        window.fill((255, 255, 255))

        # Blit the character surface onto the window
        window.blit(character_surface, (50, 50))

        # Update the display
        pygame.display.update()

    # Quit Pygame
    pygame.quit()

# Create the GUI window
window = tk.Tk()
window.title("Emotion Prediction")

# Hide the root window
window.withdraw()

# Create a pop-up dialog to get user's name
name = simpledialog.askstring("Name", "Enter your name:")

# Show the root window again
window.deiconify()

# Create a label to display the user's name
name_label = tk.Label(window, text="Hello, {}".format(name))
name_label.pack()

# Start the recording and prediction automatically
predict_emotion()

# Start the GUI event loop
window.mainloop()
```
